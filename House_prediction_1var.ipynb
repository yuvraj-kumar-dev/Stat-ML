```python
#House_Prediction using one independent variable (Simple Linear Regression)

from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
df = pd.read_csv('House_price.csv')
df.head()
# df.keys()
```

```python
model = LinearRegression()
model.fit(df[['Area_in_m2']], df['Price_in_L'])
```

```python
model.predict([[1500]])
```
