```python
import numpy as np
import pandas as pd
```

```python
def gradient_descent(x, y, lr=0.01, epochs=3000):
    m,b = 0.0, 0.0
    
    for epoch in range(epochs):
        y_pred = m*x + b
        error = y - y_pred
        cost = np.mean(error**2)
        dm = -2*np.mean(x*error)
        db = -2*np.mean(error)
        b = b - lr*db
        m = m - lr*dm

        print(f"m = {m}, b={b}, epoch={epoch}, cost={cost}")
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

#y = 2x+3

gradient_descent(x,y) #m=2 and b = 2.9999 approx 3
```
