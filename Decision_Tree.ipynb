```python
import pandas as pd

df = pd.read_csv('Salaries.csv')

df.head()
```

```python
df.columns.tolist()

df.columns = df.columns.str.strip()

df.Degree
```

```python
df['Degree_number'] = df.Degree.map({'Bachelor': 1, 'Masters': 2})

df.head()
```

```python
df.drop('Degree', axis='columns', inplace=True)
```

```python
df_encoded = pd.get_dummies(df, columns=['Company', 'Job'], drop_first=True)

df_encoded
```

```python
### Model Training
```

```python
from sklearn.tree import DecisionTreeClassifier

X = df_encoded.drop('Salary>100K', axis='columns')
y = df_encoded['Salary>100K']

model = DecisionTreeClassifier(criterion='entropy') #by default criterion='gini'
model.fit(X, y)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X)
report = classification_report(y, y_pred)

print(report)
```

```python
from sklearn.tree import export_text

print(export_text(model, feature_names=list(X.columns)))
```
