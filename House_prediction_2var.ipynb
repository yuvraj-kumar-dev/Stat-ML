```python
#House_Prediction using more than one independent variable (Multiple Linear Regression)

from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
df = pd.read_csv('House_price_2.csv')
df.sample(5)
# df.keys()
```

```python
model = LinearRegression()
model.fit(df[['Area_in_m2', 'Bedrooms']], df['Price_in_L'])
```

```python
model.predict([[1500, 3]])
```

```python
test = pd.DataFrame([
    {'Area_in_m2':1500, 'Bedrooms':2},
    {'Area_in_m2':1500, 'Bedrooms':3}
])
```

```python
model.predict(test)
```
