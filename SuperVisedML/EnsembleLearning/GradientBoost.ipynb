```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import load_diabetes

dataset = load_diabetes()
print(dataset)
```

```python
dataset['data']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(dataset['data'], dataset['target'], test_size=0.3, random_state=42)
```

### Train Linear Regressor

```python
from sklearn.metrics import r2_score
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

model.score(X_test, y_test)
```

```python
model.predict([X_test[3]])
```

```python
y_test[3]
```

### Train Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(max_depth=4, min_samples_split=10, min_samples_leaf=5, random_state=42)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

### Train GDM

```python
from sklearn.ensemble import GradientBoostingRegressor
model = GradientBoostingRegressor(n_estimators=50)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

### USING LASSO (L1) REGULARIZATION

```python
from sklearn.linear_model import Lasso, Ridge

lasso_model = Lasso()
lasso_model.fit(X_train, y_train)
lasso_model.score(X_test, y_test)
```

```python
lasso_model.coef_
```

### using ridge (L2) regularization

```python
ridge = Ridge()
ridge.fit(X_train, y_train)
ridge.score(X_test, y_test)
```

### using RandomForest

```python
from sklearn.ensemble import RandomForestRegressor

randomForest = RandomForestRegressor()
randomForest.fit(X_train, y_train)
randomForest.score(X_test, y_test)
```

### Using XGBRegressor

```python
from xgboost import XGBRegressor

model=XGBRegressor(n_estimators=400, learning_rate=0.05, max_depth=3, reg_alpha=1.0, random_state=42)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```
